```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# seed for reproducing
np.random.seed(19680801)

# Example data
features = {}
for i in range(1,15):
    features["issueFeature" + str(i)] = np.random.rand()
for i in range(1, 7):
    features["sprintFeature" + str(i)] = np.random.rand()
y_pos = np.arange(len(features))
error = 0.1 * np.random.rand(len(features))

ax.barh(y_pos, features.values(), xerr=error, align='center')
ax.set_yticks(y_pos, labels=features)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('Feature importance')

plt.show()
```
